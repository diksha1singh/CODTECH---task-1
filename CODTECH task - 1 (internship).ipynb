# Exploratory Data Analysis (EDA)

### Load the necessary libraries and dataset

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Iris dataset from seaborn's built-in datasets
df = sns.load_dataset('iris')
```

```python
# Display the first few rows of the dataset
df.head()  
```

### Explore the basic characteristics of the dataset

```python
# Display the shape of the dataset
print(f"Shape of the dataset: {df.shape}")
```

```python
# Display the data types of each column
print("\nData types of each column:")
print(df.dtypes)
```

```python
# Display summary statistics
print("\nSummary statistics:")
print(df.describe())  
```

### Check for missing values

```python
# Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())   
```

### Visualize the distribution of each feature

```python
# Plot histograms for each numerical feature
df.hist(bins=20, figsize=(10, 8))
plt.suptitle("Histograms of Numerical Features")
plt.show()
```

```python
# Plot box plots for each numerical feature
plt.figure(figsize=(10, 8))
sns.boxplot(data=df)
plt.title("Box Plots of Numerical Features")
plt.show()
```

```python

# Scatter plot of sepal_length vs sepal_width
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.title("Sepal Length vs Sepal Width")
plt.show()  
```

```python
# Pair plot of all numerical features
sns.pairplot(df, hue='species')
plt.suptitle("Pair Plot of Numerical Features", y=1.02)
plt.show()
```

## Correlation analysis

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()
```

```python
# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

## Identify and visualize outliers

```python
# Identify and visualize outliers using box plots
plt.figure(figsize=(7, 5))
sns.boxplot(data=df)
plt.title("Box Plots to Identify Outliers")
plt.show()
```

# Summary of Findings:
### The dataset contains 150 observations and 5 columns (4 numerical features and 1 categorical feature).
### There are no missing values in the dataset.
### The summary statistics indicate that the numerical features have varying ranges and distributions.
### The histograms and box plots reveal that some features have skewed distributions and potential outliers.
### The scatter plots and pair plots show that there are relationships between some numerical features, and these relationships vary across different species.
### The correlation heatmap indicates that some features are moderately correlated.
